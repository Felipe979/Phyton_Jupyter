# Ajuste de Curvas e Análise de Regressão

```python
import numpy as np
x = np.array([3.4, 4.6, 5.2, 8.0, 10.7])
y = np.array([48, 52, 58, 76, 96])

mod_linear = np.polyfit(x, y, 1)

a = float(mod_linear[0])
b = float(mod_linear[1])
y_mod = a * x + b

def lmod(a, b, x):
    '''
    Retorna o valor de uma função linear do tipo y = ax + b, onde 'a' é o coeficiente angular,
    'b' é o coeficiente linear e 'x' é o ponto onde se deseja calcular a função.
    '''
    return a * x + b

import matplotlib.pyplot as plt
plt.style.use('seaborn') 
plt.figure(dpi=200)
plt.plot(x, y, 'o', color = 'm', label = 'Pontos Experimentais')
#plt.plot(x, y_mod, color = 'b', label = 'Reta Ajustada' )
plt.xlabel(r'$\mathrm{\mathbf{Unidades\ de\ Poluente}}$')
plt.xticks(np.array(np.arange(2,12,1)))
plt.ylabel(r'$\mathrm{\mathbf{Incidência\ da\ Doença}}$')
plt.yticks(np.array(np.arange(30,101,10))) #Altera o espaçamento entre as malhas do grid vertical
plt.title(r'$\mathrm{\mathbf{Poluente \times Patologia}}$')
plt.grid(color='black', linewidth=0.5)
plt.axis([2, 12, 30, 100])
plt.legend(frameon = True, framealpha=0.7, facecolor = 'white', loc='upper left')
#plt.show()
plt.savefig(dpi=200,fname='pontos2.svg')
```

```python
# Criando um intervalo para plotar a reta modelo que melhor representa
# o comportamento do fenômeno

x1 = np.array(np.arange(0,13,1))
y_mod1 = a * x1 + b

plt.style.use('seaborn') 
plt.figure(dpi=200)
plt.plot(x, y, 'o', color = 'm', label = 'Pontos Experimentais')
plt.plot(x1, y_mod1, color = 'b', label = 'Reta Ajustada' )
plt.xlabel(r'$\mathrm{\mathbf{Unidades\ de\ Poluente}}$')
plt.xticks(np.array(np.arange(3,12,1)))
plt.ylabel(r'$\mathrm{\mathbf{Incidência\ da\ Doença}}$')
plt.yticks(np.array(np.arange(30,101,10)))
plt.title(r'$\mathrm{\mathbf{Poluente \times Patologia}}$')
plt.grid(color='black', linewidth=0.5)
plt.axis([3, 12, 30, 100])
plt.legend(frameon = True, framealpha=0.7, facecolor = 'white', loc='upper left')
#plt.show()
plt.savefig(dpi=200,fname='curva ajustada2.svg')
```

```python
l = lmod(a,b,7.3)
print("y = {:.4f}x + {:.4f}\n".format(a,b))
print("y(7.3) = {:.4f}".format(l))
```
